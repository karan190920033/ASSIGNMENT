```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.figure(figsize = (10, 6))
n = np.arange(1, 1e3)
plt.plot(np.log(n), label = 'log(n)')
plt.plot(n, label = 'n')
plt.plot(n**2, label = '$n^2$')
plt.yscale('log')
plt.legend()
plt.show()
```
